```python
#call library
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
#text preprocessing (text mining)
#list of stopwords
stop=stopwords.words("english") #stop user defined object
```

```python
#create user defined function for text preprocessing
def clean_text(msg):#clean_text() user defined function
  #step1 : first convert all msg into lower case
  msg=msg.lower()
  #step2:apply tokenizer
  tokens=word_tokenize(msg)
  #filtering only alphabets , means to remove digit and special
  #characters(punctuation marks) then use inbuilt method isalpha()
  word_token=[w for w in tokens if w.isalpha()]
  #Remove Stopwords
  #use list comprehension
  clean_token=[w for w in word_token if w not in stop]
  #next steps of preprocessing : Lemmatization
  #Create the object of WordNetLemmatizer class
  lemma=WordNetLemmatizer()
  clean_token=[lemma.lemmatize(w) for w in clean_token]
  #lemmatize() inbuilt method of WordNetLemmatizer class
  return' '.join(clean_token) #to make it a sentence
  '''print(tokens)
  print("after remove digit and special character")
  print(word_token)
  print("After remove stopwords")
  print(clean_token)'''
```

```python
#call function means we apply the clean_text() on the entire column
#of message of dataset
df["Message"]=df["Message"].apply(clean_text)
df["Message"].head()
```

```python
#select input and output
x=df["Message"]#input
y=df["Category"]#output
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=1,test_size=0.3)
```

```python
xtrain.shape,xtest.shape
```

```python
#Now once we have the clean text in hand, The next thing that we would like to do is convert the text
#into vector/number.

#First , we trying the first technique : - CountVectorizer() :
# to converts text into vector/number

#first call class CountVectorizer inbuilt class
from sklearn.feature_extraction.text import CountVectorizer

#Create the object of CountVectorizer class
cv=CountVectorizer()
```

```python
#apply CountVectorizer on xtrain and xtest
#use fit_transform() inbuilt method of CountVectorizer on xtrain
#use fit_transform() inbuilt method of CountVectorizer on xtest
xtrain_cv=cv.fit_transform(xtrain)
xtest_cv=cv.transform(xtest)
```

```python
#to check feature extract from message after apply CountVectorizer
cv.get_feature_names_out()
```

```python
xtrain_cv.toarray()
```

```python
pd.DataFrame(xtrain_cv.toarray(),columns=cv.get_feature_names_out())
```

```python
xtrain_cv.shape
```

```python
pd.DataFrame(xtest_cv.toarray(),columns=cv.get_feature_names_out())
```

```python
#create a model with DecisionTreeClassifier Algorithm
from sklearn.tree import DecisionTreeClassifier
#create object of DecisionTreeClassifier class
dt=DecisionTreeClassifier(random_state=1)
```

```python
#Train the model with 70% data
dt.fit(xtrain_cv,ytrain)
```

```python
#test the model with 30%
ypred=dt.predict(xtest_cv)
```

```python
#generate Report
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(ytest,ypred))
```
